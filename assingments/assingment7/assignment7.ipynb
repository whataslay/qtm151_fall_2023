(c) Clean a column

<font size = "3">

- Our goal is to replace any string values in the column "position" <br>
to missing values (NaNs) and covert to numeric

- To do so
    1. Extract list of non-numeric values
    2. Replace certain list of values
    3. Convert column to numeric
    4. Display


```python
# Write your own code here
# Create and store a list of non-numeric values

nonnumeric = list(sprint_results.query("position.str.isnumeric()==False")["position"]) #nonnumeric values
replacement = [np.nan]*len(nonnumeric) #create list for replacements
sprint_results["position"] = sprint_results["position"].replace(nonnumeric,replacement) #replacing it on the df
sprint_results["position"] = pd.to_numeric(sprint_results["position"]) #converting it to numberic
display(sprint_results)
```

(d) Groupby + Aggregate

<font size = "3">

- Compute the mean and standard deviation of "position", <br>
grouping by "constructorId". Store the new dataset <br>
as "df_aggConstrcId_pos"

- Sort the aggregate dataset in ascending <br>
order of mean position using ".sort_values()" <br>
(i.e., to find the constructor with drivers having the best mean positions)

- Display

- Print which constructorId has the lowest mean position.

HINT: See Lecture 13 <br>


```python
# Write your own code
df_aggConstrcld_pos = (sprint_results.groupby("constructorId")
                       .agg(mean_position=('position','mean'),
                        std_position=('position','std'))) #grouping by constructor ID
display (df_aggConstrcld_pos.sort_values("mean_position",ascending=True)) #displaying the whole sorted by mean_position
df_aggConstrcld_pos.sort_values("mean_position",ascending=True)[:1] #getting the smallest mean
```

(d) Query + Groupby + Aggregate

<font size = "3">

- Use ".query()" to subset observations with <br>

- Compute the mean and standard deviation of "position", <br>

- Display

Hint: See Lectures 9 and 11 (to refresh ".query()") and <br>

```python
# Write your own code
df_subAggConstr_pos = (sprint_results.query("raceId>=100")
                        .groupby("constructorId")
                        .agg(mean_position = ("position","mean"),
                        std_position = ("position","std")))

display (df_subAggConstr_pos)
```
