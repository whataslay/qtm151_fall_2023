- Import the datset "data/bmi.csv"

- Write a lambda function for $ BMI = \frac{weight (kg)}{height (m)^2} $
- Use ```list(map(...))``` to calculate $BMI$ for each row of the dataset
- Round the outcome into 2 decimal points
- Store the calculation in a new column in the dataframe called `BMI`
- Display the dataframe


````python
bmidataset = pd.read_csv("data/bmi.csv") #import dataset
bmifunc = lambda weightinkg,heightinm: weightinkg/(heightinm**2) #making the bmi lambda function
bmilist = list(map(bmifunc,bmidataset["Weight"],bmidataset["Height"]/100)) #making a bmi list from height (divided by 100 to turn cm to m) and weight 
bmilist = list(map(lambda num:round (num,2),bmilist)) #bmilist is rounded to 2 decimal points
bmidataset["BMI"]=bmilist #store the calculation in a new col called BM
display(bmidataset) #display the dataframe
````

(e) Plot by category

<font size = "3">

- Read "data/wdi_urban.csv"<br>
This contains data from 1980 and 2020

- Create a list with the unique values of "year"
- Create a plot with a ...
    - Histogram of "prop_urbanpopulation" <br>
    for each year
    - Set the histogram option ``` alpha = 0.5 ``` for transparency
    - All the histograms should be overlapping
    - Create a legend with the list of years
    - Label the axes and the title

What do we learn from the graph?

HINT: See the example at the end of Lecture 10


````python
# Write your own code here
wdiurban = pd.read_csv("data/wdi_urban.csv")
listofyrs = pd.unique(wdiurban["year"])

for year in listofyrs:
    thisyearprop =  wdiurban.query("year==@year")
    plt.hist(x=thisyearprop["prop_urbanpopulation"],alpha=0.5)
plt.legend(listofyrs)
plt.xlabel("prop_urbanpopulation")
plt.ylabel("freq")
plt.title("prop_urbanpopulation distribution for 1980 and 2020")

#From this graph, we can see that the proportion of urban population was generally higher in 2020 than 1980. Similarily, the range of urban proportion distributuion for 2020 has shrunk and didn't have the low values we see for 1980
````
